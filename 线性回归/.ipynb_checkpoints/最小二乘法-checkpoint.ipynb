```python
import numpy as np
import pandas as pd
```

#### 波士顿房价数据集字段说明
* CRIM 房屋所在镇的犯罪率
* ZN 面积大于25000平方英尺住宅所占的比例
* INDUS 房屋所在镇非零售区域所占比例
* CHAS 房屋是否位于河边，在河边则为1 ，否则为0
* NOX 一氧化氮的浓度
* RM 平均房屋数量
* AGE 1940年前建成房屋所占的比例
* DIS 房屋距离波士顿五大就业中心的加权距离
* RAD 距离房屋最近的公路
* TAX 财产税额度
* PTRATIO 房屋所在镇的师生比例
* B 计算公式 1000* （房屋所在镇非美籍人口所在比例 - 0.63） ** 2
* LSTAT 弱势群体人口所在比例
* MEDV 房屋的平均价格


```python
data = pd.read_csv(r"boston.csv")
# 查看数据基本信息，也可以来看是否存在缺失值
# data.info()
# 查看是否存在重复值
# data.duplicated().any()
```

```python
class LinearRegression:
    ''' 实现线性回归（最小二乘法）
    '''
    def fit(self, X, y):
        ''' 根据提供的训练数据X，对模型进行训练
        Parameters 
        -------
        X: 类数组类型 特征矩阵 形状[样本数量，特征数量]
        y: 类数组类型 [样本数量]
        
        '''
        # X 要求是一个完整的矩阵
        # 说明： 如果X 是数组对象的一部分，而不是完整的对象数据（例 X是由其他对象通过切片传递过来的）
        # 则无法完成矩阵的转换
        # 创建X的拷贝对象，避免转换矩阵时失败
        X = np.asmatrix(X.copy())
        # y 是一维结构（行或列向量），可以不进行拷贝
        y = np.asmatrix(y).reshape(-1, 1)
        # 通过最小二乘公式，求值
        self.w_ = (X.T * X).I * X.T * y
    
    def predict(self,X):
        ''' 根据参数传递的样本 X 对样本数据进行预测
        Parameters 
        -------
        X: 类数组类型 特征矩阵 形状[样本数量，特征数量]
           待预测的样本特征（属性）
        
        Returns
        ---
        result: 数组类型 预测的结果
        '''
        # 说明： 如果X 是数组对象的一部分，而不是完整的对象数据（例 X是由其他对象通过切片传递过来的）
        # 则无法完成矩阵的转换
        # 创建X的拷贝对象，避免转换矩阵时失败
        X = np.asmatrix(X.copy())
        result = X * self.w_
        # 将矩阵转换成 ndarray 数组，进行扁平化(ravel)处理后返回结果
        return np.array(result).ravel()
        
        
```

```python
# 不考虑截距的情况
t = data.sample(len(data),random_state = 0)
train_X = t.iloc[:400, :-1]
train_y = t.iloc[:400, -1]
test_X = t.iloc[400:, :-1]
test_y = t.iloc[400:, -1]

lr = LinearRegression()
lr.fit(train_X, train_y)
result = lr.predict(test_X)
display(np.mean((result - test_y) ** 2))
# display(lr.w_)
```

```python
# 考虑截距的情况，增加一列，让该列的所有值均为1
t = data.sample(len(data),random_state = 0)
# 增加一列
# t["Intercept"] = 1 
# 按照习惯截距作为 W0 为之新增 X0 列置于最前面
new_columns = t.columns.insert(0,"Intercept")
# 重新安排列的顺序，若值为空，则使用 fill_value  参数指定的值进行填充
t1  = t.reindex(columns = new_columns,fill_value = 1)
train_X1 = t1.iloc[:400, :-1]
train_y1 = t1.iloc[:400, -1]
test_X1 = t1.iloc[400:, :-1]
test_y1 = t1.iloc[400:, -1]

lr = LinearRegression()
lr.fit(train_X1, train_y1)
result1 = lr.predict(test_X1)
display(np.mean((result1 - test_y1) ** 2))
# display(lr.w_)
```
